```python
import numpy as np
import pandas as pd
```

```python
import ast
ast.literal_eval
def convert(obj):
    L =[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        return L
```

```python
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies.head()
```

```python
#cast ko bhi change krna hai
# import ast

# def convert3(obj):
#     L = []
#     counter = 0
#     # Safely evaluate the string as a Python object
#     for i in ast.literal_eval(obj):
#         if counter != 3:
#             L.append(i['name'])
#             counter += 1
#         else:
#             break
#     return L
import ast

def convert3(obj):
    # If it's already a list, use it directly
    if isinstance(obj, list):
        data = obj
    else:
        # Safely convert string to Python object
        data = ast.literal_eval(obj)

    # Ensure we only take up to 3 names
    return [i['name'] if isinstance(i, dict) else i for i in data[:3]]
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
# movies['genres'] = movies['genres'].apply(
#     lambda X: [i.replace(" ", "") for i in X] 
# )
movies['cast'] = movies['cast'].apply(
    lambda X: [i.replace(" ", "") for i in X])
movies['crew'] = movies['crew'].apply(
    lambda X: [i.replace(" ", "") for i in X]
)
# movies['keywords'] = movies['keywords'].apply(
    # lambda X: [i.replace(" ", "") for i in X])

```

```python
movies['genres'] = movies['genres'].apply(
    lambda X: [i.replace(" ", "") for i in X]if isinstance(X, list) else []
)
```

```python
movies['keywords'] = movies['keywords'].apply(
    lambda X: [i.replace(" ", "") for i in X]if isinstance(X, list) else []
)
movies.head()
```

```python
#form tag 
movies['tag']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tag']]
```

```python
new_df
```

```python
new_df['tag']=new_df['tag'].apply(lambda X:" ".join(X))
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tag']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
# use stem for loved,love ,loving into love 
# !pip install nltk
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
      y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df=new_df['tag'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda X: X[1])[1:6]
```

```python
def recommend(movie):
    movies_index = new_df[new_df['title'] == movie].index[0]
    distance=similarity[movies_index]
    movies_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x: x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
    
```

```python
recommend('Avatar')
```
